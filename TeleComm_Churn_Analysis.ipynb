```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data (use your actual DataFrame)
# df = pd.DataFrame(...)

# Calculate the counts for each group and normalize them
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()
percent_data = count_data.apply(lambda x: x / x.sum() * 100, axis=1)

# Plot stacked bar chart
plt.figure(figsize = (4,4))
percent_data.plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e']) # Customize colors if needed
plt.title("Percentage of Churn by Senior Citizen")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.legend(title="Churn", loc="upper left", bbox_to_anchor=(1, 1))

# Add percentages on top of each segment
for i in range(percent_data.shape[0]):
    for j in range(percent_data.shape[1]):
        plt.text(i, percent_data.iloc[i, :j+1].sum() - percent_data.iloc[i, j]/2, 
                 f"{percent_data.iloc[i, j]:.1f}%", ha="center", va="center", color="white")

plt.show()

```
